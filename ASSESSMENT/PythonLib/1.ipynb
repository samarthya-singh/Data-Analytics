```python
import pandas as pd
import numpy as np
```

```python
# 1) Import pandas and read in the banklist.csv file into a dataframe called
# banks.


banks=pd.read_csv("banklist.csv")
```

```python
#4) How many States (ST) are represented in this data set?

uniqe_state=banks['ST'].nunique
print("Number of unique states:", uniqe_state)
```

```python
#5) Get a list or array of all the states in the data set.

state_array=np.array(banks['ST'])
print(state_array)
```

```python
# 6) What are the top 5 states with the most failed banks?


state_counts = banks['ST'].value_counts()

top5_states = state_counts.head(5)

top5_states
```

```python
# 7) What are the top 5 acquiring institutions?

acquiring_count=banks['Acquiring Institution'].value_counts()

top5_acquiring=acquiring_count.head()

top5_acquiring
```

```python

# 8) How many banks has the State Bank of Texas acquired? How many of
# them were actually in Texas?


# Filter banks acquired by State Bank of Texas
acquired_by_sbt = banks[banks['Acquiring Institution'] == 'State Bank of Texas']
# Total number of banks acquired
total_acquired = len(acquired_by_sbt)
print("Total banks acquired by State Bank of Texas:", total_acquired)

# Number of banks located in Texas (ST == 'TX')
in_texas = len(acquired_by_sbt[acquired_by_sbt['ST'] == 'TX'])
print("Banks acquired that were actually in Texas:", in_texas)
```

```python
# 9) What is the most common city in California for a bank to fail in?

# Filter only banks in California
ca_banks = banks[banks['ST'] == 'CA']

# Count number of failed banks per city
city_counts = ca_banks['City'].value_counts()

# Get the most common city
most_common_city = city_counts.idxmax()   # or use .head(1)
print("Most common city in California for a bank to fail:", most_common_city)

```
